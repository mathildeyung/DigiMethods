## 5.1.1 Basic Graph Types


```python
# Basic visualisations

# Visualisations can have these potential components --> axes, ticks (points, marks), labels, titles, colors,
# facets (e.g. splitting up groups), legends (what everything means/stands for). Often each tick will represent 
# several observations - as such you should consider if and how you would want to represent this in your plot.

# A very basic type of visualisations is line plots. You have to be mindful of the origins of you plot (e.g. does 
# the axes start at 0 or something else). You also have to think about the original data - is the visualisation 
# showing a trend that looks really strong because of the way it is plotted? 
# Other types are bar graphs, scatter plots, and histograms. 

# The visualisations should be as self-contained (should be readable without going through every note/text), 
# accessible (are the readable to everyone? e.g. colour, size - look at colorbrewer 2.0), 
# and intuitive as possible. 
# Ink-to-information ratio - don't add anything that is unecessary. It should always be as easy to read as 
# possible. 
```

## 5.1.2 Formatting Graphs

Adding a title, axis, and tick labels for comprehension

```python
values = [item[1] for item in flu_death_rates]
labels = [item[0] for item in flu_death_rates]

plt.bar(labels, values)

#commands for adding title, axis etc. 

plt.title("Bar chart to show US 2016 flu/pneumonia death rates by state")
plt.xlabel("States")
plt.ylabel("Deaths per 100,000")
plt.xticks(rotation=90)

plt.tight_layout() # to fit it to the figure size
```

Adjusting the plot size

```python
values = [item[1] for item in flu_death_rates]
labels = [item[0] for item in flu_death_rates]

plt.figure(figsize=(10, 7), dpi= 120)

plt.bar(labels, values)

plt.title("Bar chart to show US 2016 flu/pneumonia death rates by state")
plt.xlabel("States")
plt.ylabel("Deaths per 100,000")
plt.xticks(rotation=90) #so they don't overlap

plt.tight_layout()
```
