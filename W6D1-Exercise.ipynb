**Exercise 6.1.1.1** Read in the dataset "nchs_death_causes.csv" using `pandas` methods.

```python
import pandas as pd
```

```python
us_deaths_2016 = pd.read_csv('nchs_death_causes.csv')

us_deaths_2016
```

**Exercise 6.1.1.2** Make a `DataFrame` that is a subset of the data, containing only one cause of death and one state.

```python
all_stroke = us_deaths_2016[us_deaths_2016['Cause Name'].isin(['Stroke'])]

california_stroke = all_stroke[all_stroke['State'].isin(['California'])]
```

```python
california_stroke_df = pd.DataFrame(california_stroke)
```

```python
california_stroke_df
```

**Exercise 6.1.1.4** Make your figure self-contained and export it/save it to your machine.

**Exercise 6.1.1.5** Make a `DataFrame` that is a subset of the data, containing _two_ causes of death that might be interesting to compare for one state.

```python
california_deaths = us_deaths_2016[us_deaths_2016['State'].isin(['California'])]

california_deaths_df = pd.DataFrame(california_deaths)

california_deaths_df
```

```python
two_causes_california = california_deaths_df[california_deaths_df['Cause Name'].isin(['Stroke','Diabetes'])]

two_causes_california_df = pd.DataFrame(two_causes_california)

two_causes_california_df
```

**Exercise 6.1.1.6** Create a _self-contained_ scatter plot that compares the two causes of death in the state.

```python
plt.plot(two_causes_california_df['Deaths'][two_causes_california_df['Cause Name'] == 'Stroke'],
         two_causes_california_df['Deaths'][two_causes_california_df['Cause Name'] == 'Diabetes'], 'o')


plt.title('Deaths caused by stroke and diabetes in California')
plt.xlabel("Stroke")
plt.ylabel("Diabetes")
#plt.xticks(rotation=90)

plt.show ()

#file_dir = os.getcwd()

#plt.savefig(file_dir + "/Stroke_Deaths_California.png")
```

**Exercise 6.1.1.7** Make a `DataFrame` that is a subset of the data, containing data for only one year and one state.

```python
#california_deaths_2004 =(california_deaths_df['Year'] == '2004')

california_deaths_2004 = california_deaths_df[california_deaths_df['Year'].isin(['2004'])]

california_deaths_2004
```

**Exercise 6.1.1.8** Create a  _self-contained_ bar graph showing the death rate across causes of death in your subset of the data.

```python
plt.figure(figsize=(10, 7), dpi= 120)

plt.bar(california_deaths_2004['Cause Name'], california_deaths_2004['Deaths'])


plt.title("Bar chart that shows number of deaths by cause in California 2004")
plt.xlabel("Cause")
plt.ylabel("Deaths per 100,000")
plt.xticks(rotation = 90)
```
